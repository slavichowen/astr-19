```python
import numpy
import matplotlib.pyplot as plt
```

```python
rngseed = numpy.random.default_rng()
data = rngseed.uniform(0.0,1.0,1000)
```

```python
plt.figure()
plt.hist(data, bins=100)
plt.xlabel("value")
plt.ylabel("count")
plt.title("1000 random numbers between 0 and 1, 100 bins")
plt.savefig("entry 3 graph", format="pdf", bbox_inches="tight")
plt.show()
```
