```python
import numpy
import matplotlib.pyplot as plt
```

```python
x = numpy.linspace(0,1,100)

def sin(x):
    return numpy.sin(x)

def cos(t):
    return numpy.cos(x)
```

```python
sin_y = sin(x)
cos_y = cos(x)

f, a = plt.subplots(1,2,figsize=(10,4))

a[0].plot(x,sin_y)
a[0].set_title("sin(x)")
a[0].set_xlabel("x")
a[0].set_ylabel("sin(x)")

a[1].plot(x,cos_y)
a[1].set_title("cos(x)")
a[1].set_xlabel("x")
a[1].set_ylabel("cos(x)")

f.tight_layout()
f.savefig("session8graph", format="pdf", bbox_inches="tight")
plt.show()
```
