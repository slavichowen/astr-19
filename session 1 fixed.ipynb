```python
import numpy
import matplotlib.pyplot as plt
```

```python
x = numpy.linspace(0,1,100)

def exp(x):
    return numpy.exp(x)
    
```

```python
y = exp(x)

plt.figure()
plt.plot(x,y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("x vs. exp(x)")
plt.savefig("session7_exp_plot.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
