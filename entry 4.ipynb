```python
import numpy
import matplotlib.pyplot as plt
```

```python
rngseed = numpy.random.default_rng()
data = rngseed.normal(loc=0.0,scale=1.0,size=1000)
```

```python
plt.figure()
plt.hist(data,bins=100)
plt.xlabel("value")
plt.ylabel("count")
plt.title("normal distr")
plt.savefig("entry 4 graph.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
